# DataSet : Indira_Gandhi_old_aged_pension_scheme

# Dataset Description

The dataset is about the Indira Gandhi Old Age Pension Scheme (IGNOAPS), a social welfare 
program in India. The scheme provides financial assistance to elderly individuals belonging
to economically weaker sections, including Scheduled Castes (SC), Scheduled Tribes (ST),
Other Backward Classes (OBC), and the General category. The dataset includes details on 
the number of beneficiaries, Aadhaar-linked accounts, and mobile registrations across 
various states and districts. This dataset can be used for policy evaluation, beneficiary
analysis, and scheme performance tracking in India.

# Dataset Overview

The fin_year column represents the financial year of the record, such as "2024-2025." The mnth column indicates the month in numerical format.

The lgd_state_code is a unique identifier for each state, while state_name provides the name of the state. Similarly, lgd_district_code is the unique identifier for districts, and district_name represents the districtâ€™s name.

The scheme_code column specifies the scheme under which the beneficiaries are covered, such as "IGNOAPS." The total_beneficiaries column represents the total number of beneficiaries in a district.

The dataset also categorizes beneficiaries based on their social background. total_sc denotes the number of Scheduled Caste (SC) beneficiaries, total_st represents the Scheduled Tribe (ST) beneficiaries, total_gen indicates the General category beneficiaries, and total_obc includes beneficiaries from Other Backward Classes (OBC).

Additional details include total_aadhar, which refers to the number of beneficiaries with Aadhaar-linked accounts, and total_mobileno, which represents the number of beneficiaries with registered mobile numbers.

Lastly, the lastUpdated column records the date when the data was last updated.

```python
beneficiary_categories = {
    "SC": df["total_sc"].sum(),
    "ST": df["total_st"].sum(),
    "General": df["total_gen"].sum(),
    "OBC": df["total_obc"].sum(),
}

plt.figure(figsize=(8, 6))
colors = ["#FF9999", "#66B3FF", "#99FF99", "#FFD700"]
plt.pie(beneficiary_categories.values(), labels=beneficiary_categories.keys(), autopct="%1.1f%%", colors=colors, startangle=140)
plt.title("Distribution of Beneficiaries by Category")
plt.show()
```

# Observations from Data Analysis:


Beneficiary Distribution Variability: The number of beneficiaries varies significantly across states and districts. Some districts have a high concentration of beneficiaries, while others have relatively fewer recipients.

Caste-Wise Beneficiary Trends: The distribution of beneficiaries among SC, ST, OBC, and General categories is uneven. Some districts have a dominant caste group receiving benefits, while others have a more balanced distribution.

Aadhar & Mobile Number Linkage: There is a notable difference in Aadhar and mobile number linkage across districts. Some areas have near-complete Aadhar verification, while others have low registration rates.

State-Wise Differences: Certain states have a significantly higher number of beneficiaries compared to others. This could be due to demographic factors or differences in implementation efficiency.

Monthly Trends: The dataset includes data for multiple months, allowing for analysis of how the number of beneficiaries changes over time. Some months might see a rise or fall in new enrollments.

Potential Areas for Further Analysis 

State & District-Wise Trends: Identify which states and districts have the highest and lowest beneficiary counts.

Aadhar & Mobile Penetration: Examine why some districts have low Aadhar or mobile number linkage and whether this affects pension distribution.

Growth Over Time: Analyze trends in beneficiary numbers across financial years to detect increasing or decreasing participation.

Efficiency of Fund Utilization: Compare districts with high vs. low Aadhar-linked beneficiaries to assess how digital verification impacts pension disbursement.
